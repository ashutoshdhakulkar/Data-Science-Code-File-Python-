# Scaling Numerical Features:

Standard Scaling:

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# standard scaling
standard_scaler = StandardScaler()
scaled_df = df.copy()
scaled_df[df.select_dtypes(include=['int64','float64']).columns]=standard_scaler.fit_transform(df.select_dtypes(include=['int64', 'float64']))

```

```python
# MinMaxScaling
minmax_scaler = MinMaxScaler()
minmax_scaled_df =df.copy()
minmax_scaled_df[df.select_dtypes(include=['int64', 'float64']).columns] = minmax_scaler.fit_transform(df.select_dtypes(include=['int64', 'float64']))
```

Standard Scaling is preferred when features follow a normal distribution and are to be used in linear models.

Min-Max Scaling is preferred when the data does not follow a normal distribution and is to be used in algorithms that do not assume any distribution, such as neural networks.

# Encoding Techniques
One-Hot Encoding and Label Encoding

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
df['workclass']= label_encoder.fit_transform(df['workclass'])
df['education']= label_encoder.fit_transform(df['education'])
df['marital_status']= label_encoder.fit_transform(df['marital_status'])
df['occupation']= label_encoder.fit_transform(df['occupation'])
df['relationship']= label_encoder.fit_transform(df['relationship'])
df['race']= label_encoder.fit_transform(df['race'])
df['sex']= label_encoder.fit_transform(df['sex'])
df['native_country']= label_encoder.fit_transform(df['native_country'])
```

# One-Hot Encoding:
 Does not assume any order, works well with algorithms expecting categorical inputs.
 Can lead to high dimensionality if there are many categories.

# Label Encoding :
Simple and fast, keeps dimensionality low.
Assumes an ordinal relationship, which may not be appropriate for all categorical variables.

# **Feature Engineering**

```python
# Creating two new features to be added

df['age_binned'] = pd.cut(df['age'], bins=[0, 18, 35, 60, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])

df['hours_per_week_category'] = pd.cut(df['hours_per_week'], bins=[0, 20, 40, 60, 100], labels=['Part-time', 'Full-time', 'Over-time', 'Extreme'])

print(df[['age_binned', 'hours_per_week_category']].head())
```

```python
# Transforming Skewed Feature
import numpy as np

# Log transformation
df['capital_gain'] = np.log1p(df['capital_gain'])


```

```python
print(df['capital_gain'].describe())
```

# Feature Selection

Isolation forest for Outlier Detection

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = iso_forest.fit_predict(df.select_dtypes(include=['int64', 'float64']))


```

```python
df_cleaned = df[outliers == 1]

print(f"Original shape: {df.shape}, Cleaned shape: {df_cleaned.shape}")
```

# Predictive Power Score (PPS) Analysis

```python
import ppscore as pps


pps_matrix = pps.matrix(df_cleaned)
print(pps_matrix[(pps_matrix['x'] != pps_matrix['y'])].sort_values(by='ppscore', ascending=False))


```

```python
# Compare with correlation matrix
corr_matrix = df_cleaned.corr()

print(corr_matrix)
```

* TOP FEATURES:

PPS shows capital-gain, capital-loss, education-num, hours-per-week, and age as the top features predicting income.

Correlation shows moderate linear relationships between features like capital-gain and capital-loss, hours-per-week and education-num, and age with several other features.

* INCOME PREDICTION:

PPS highlights the predictive power of capital-gain and capital-loss for predicting income, which may not show up as strongly in the correlation matrix due to non-linear relationships.

Correlation does not show income directly, as it's typically not included in feature-to-feature correlations, but it helps understand feature interactions.

# **Split the dataset into training and testing sets**

```python
import pandas as pd
df = pd.read_csv("/content/mushroom.csv")

# Display the original columns
print("Original columns order:")
print(df.columns.tolist())

# Reorder columns:
cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index('class')))
df = df[cols]
# Display the new columns order
print("\nNew columns order:")
print(df.columns.tolist())
# Display the first few rows of the reordered DataFrame
print("\nReordered DataFrame:")
print(df.head())
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder() # convert categorical values to numerical    # of all( multiple ) columns which are categorical:
df[cat_cols] = oe.fit_transform(df[cat_cols])
```

```python
from sklearn.svm import SVC
```

```python
df.head()
```

```python
x = df.iloc[:,1:]
y = df['class']
```

```python
x
```

```python
y
```

# **SVM Implementation**

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
svc = SVC() # use default values because data is in good format
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.iloc[:,1] = le.fit_transform(df.iloc[:,1])
```

```python
df['class'].value_counts()
```

```python
ytrain = le.fit_transform(ytrain)
```

```python
df['class'].unique()
```

```python
svc.fit(xtrain,ytrain)
```

```python
ypred = svc.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import classification_report
```

```python
ytest = le.fit_transform(ytest)
```

```python
print(classification_report(ytest,ypred))
```

```python
print(svc.score(xtrain,ytrain),svc.score(xtest,ytest))
```

# **Hyper-Parameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C' : [0.1,0.001,1,2,3,4,5],
    'kernel' : ['linear','poly','rbf'],
    'gamma' : [0.1,0.001,1,2,5]
}
```

```python
grid = GridSearchCV(SVC(),param_grid=params,verbose=1)
```

```python
grid.fit(xtrain,ytrain)
```

```python
xtrain
```

```python
# check optimal value of c gamma and kernel
grid.best_params_
```

# Comapring SVM performance with Various Kernals

```python
svc = SVC(C=0.1,gamma=0.1,kernel='poly')
```

```python
svc = SVC(C=0.1,gamma=0.1,kernel='linear')
```

```python
svc = SVC(C=0.1,gamma=0.1,kernel='rbf')
```

```python
svc.fit(xtrain,ytrain)
ypred = svc.predict(xtest)
```

```python
print('Training Accuracy',svc.score(xtrain,ytrain))
print('Testing Accuracy',svc.score(xtest,ytest))
```

# **Strengths:**

*  SVM is effective for high-dimensional data.
*  Regularization helps avoid overfitting.
*  Can handle non-linear relationships with the kernel trick

# **Weakness:**
*  Computationally intensive for large datasets.
*  Difficult to interpret.
*  Requires careful preprocessing.
*  High dimensionality and sparsity due to encoding categorical data.

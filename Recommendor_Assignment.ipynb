# **Feature Extraction**

```python
# change structure of dataset so that we can compute the similarity score
anime_df = anime.pivot(index='anime_id', # row = anime_id
                      columns='name', # Col = anime name
                      values='rating')
```

```python
anime_df
```

```python
anime_df.shape
```

# **Data Pre-Processing**

```python
anime_df.fillna(0, inplace=True)
```

```python
anime_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

# **Recommendation System**

```python
user_sim = 1 - pairwise_distances(anime_df.values,metric='cosine')
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)  # storing the results in  data frame
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index = anime.anime_id.unique()
user_sim_df.columns = anime.anime_id.unique()
```
